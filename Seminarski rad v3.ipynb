## Kreiranje korpusa

```python
# instalacija neophodnih modula za izvršavanje programa
# !pip install lyricwikia
# !pip install PyLyrics

# unos svih autora i naziva pesama 
from PyLyrics import *

# kreiranje korpusa za autora i listu pesama bez podataka o autoru i naslovima pesama 
def korpus_ver2(autor,pesme,ime_datoteke):

    with open(ime_datoteke, 'w') as fout: 
        for pesma in pesme: 
            fout.write('%s\n' % ( ' '.join(PyLyrics.getLyrics(autor,pesma).split())))
            fout.flush()
    return fout
    print(fout)

# definisanje parametara 
autor = 'Florence + The Machine'
pesme1 = ['Dog Days Are Over', 'Rabbit Heart (Raise It Up)', "I'm Not Calling You a Liar", 'Howl', 'Kiss with a Fist', 'Girl with One Eye', 'Drumming Song', 'Between Two Lungs', 'Cosmic Love', 'My Boy Builds Coffins', 'Hurricane Drunk', 'Blinding', "You've Got the Love", 'Bird Song', 'Falling', 'Hardest of Hearts', 'Swimming', 'Heavy in Your Arms']
pesme2 = ['Only If for a Night', 'Shake It Out', 'What the Water Gave Me', 'Never Let Me Go', 'Breaking Down', 'Lover to Lover', 'No Light, No Light', 'Seven Devils', 'Heartlines', 'Spectrum', 'All This and Heaven Too', 'Leave My Body', 'Remain Nameless', 'Strangeness & Charm', 'Bedroom Hymns', 'Landscape', 'Breath of Life']
pesme3 = ['Ship to Wreck', 'What Kind of Man', 'How Big, How Blue, How Beautiful', 'Queen of Peace', 'Various Storms & Saints', 'Delilah', 'Long & Lost', 'Caught', 'Third Eye', 'St Jude', 'Mother', 'Hiding', 'Make Up Your Mind', 'Which Witch', 'Pure Feeling', 'Conductor', 'As Far as I Could Get']
pesme4 = ['June', 'Hunger', 'South London Forever', 'Big God', 'Sky Full of Song', 'Grace', 'Patricia', '100 Years', 'The End of Love', 'No Choir']

# pozivanje funkcije kreiranja korpusa
albumi=[['lungs',pesme1],['ceremonials',pesme2],['hbhbhb',pesme3],['highashope',pesme4]]
for album in albumi:
    korpus_ver2(autor,album[1],album[0]+'4.txt')
```

## Preprocesiranje

### Tokenizacija

```python
import nltk
from nltk import sent_tokenize 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
set(stopwords.words('english'))

def tok_sent(text):
    f = open(text, 'r')
    data = str(f.readlines())
    data2 = sent_tokenize(data)
    for line in data2:
        a = line.replace("\\",'')
        print('\nRečenica:\n\n', a)
        
def tok_words(text):
    with open(text, 'r') as t:
        data = str(t.readlines())
        tokens = nltk.word_tokenize(data) 
        tokens = [token.lower() for token in tokens]
        tokens = [token for token in tokens if token.isalpha()]
        stoplist = stopwords.words('english')
        additional_stopwords = """na whoa oh ooh the ca though instead in i ah im wo ive ill hey it theres gon cause youre like to is id"""
        stoplist += additional_stopwords.split()
        tokens = [w for w in tokens if not w in stoplist]
        print('\n-------Značeće reči:\n\n', tokens)

tok_sent('ceremonials4.txt')
tok_words('ceremonials4.txt')
```

## Leksička frekvencija

```python
import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
set(stopwords.words('english'))

stoplist = stopwords.words('english')
additional_stopwords = """na whoa oh ooh the ca though instead in i ah im wo ive ill hey it theres gon cause youre like to is id"""
stoplist += additional_stopwords.split()

def most_common(album, stoplist):
    text = album+'4.txt'
    with open(text, 'r') as t:
        data = str(t.readlines())
        tokens = nltk.word_tokenize(data)
        tokens = [token.lower() for token in tokens]
        tokens = [token for token in tokens if token.isalpha()]
                    
        tokens = [w for w in tokens if not w in stoplist]
        fdist = FreqDist(tokens)
        print("NAJFREKVENTNIJE REČI:\n\n", album), fdist.plot(20,cumulative=False)
        print(fdist.most_common(100),'\n' ) 

# pozivanje funkcije prikaza najčešćih reči za jedan album
# most_common('matrix.txt', stoplist)

# prikaz najčešćih reči sa svih albuma
for album in albumi:
    most_common(album[0],stoplist)
```

## Konkordansa

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text 
import nltk.corpus  

def konkordansa(file, rec):
    with open(file, 'r') as f:
        data = str(f.readlines())
        tokens = word_tokenize(data)
        conc = nltk.Text(tokens)
        conc.concordance(rec, width = 80)
        
konkordansa('lungs4.txt', 'knife')        
```
